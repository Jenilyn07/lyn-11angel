# Flask Backend (app.py)

```python
from flask import Flask, request, jsonify

app = Flask(__name__)

# Sample data (can be updated)
place_data = {
    "location": "City Center",
    "stairs": True,
    "traffic": "High",
    "wheelchair_access": False
}

@app.route('/get_data', methods=['GET'])
def get_data():
    return jsonify(place_data)

@app.route('/update', methods=['POST'])
def update_data():
    data = request.json

    place_data["stairs"] = data.get("stairs", place_data["stairs"])
    place_data["traffic"] = data.get("traffic", place_data["traffic"])
    place_data["wheelchair_access"] = data.get("wheelchair_access", place_data["wheelchair_access"])

    return jsonify({"message": "Updated successfully", "data": place_data})

if __name__ == '__main__':
    app.run(debug=True)
```

# Frontend (index.html)

```python
<!DOCTYPE html>
<html>
<head>
    <title>Accessibility App</title>
    <style>
        body {
            font-family: Arial;
            text-align: center;
        }
        .card {
            border: 1px solid black;
            padding: 20px;
            margin: 20px;
            display: inline-block;
        }
    </style>
</head>
<body>

<h1>♿ Accessibility Monitoring System</h1>

<div class="card">
    <h3>Location Status</h3>
    <p id="stairs"></p>
    <p id="traffic"></p>
    <p id="wheelchair"></p>
</div>

<h2>Update Data</h2>

<label>Stairs:</label>
<select id="stairsInput">
    <option value="true">Yes</option>
    <option value="false">No</option>
</select>

<br><br>

<label>Traffic:</label>
<select id="trafficInput">
    <option>Low</option>
    <option>Medium</option>
    <option>High</option>
</select>

<br><br>

<label>Wheelchair Access:</label>
<select id="wheelInput">
    <option value="true">Yes</option>
    <option value="false">No</option>
</select>

<br><br>

<button onclick="updateData()">Update</button>

<script>
async function fetchData() {
    let res = await fetch("http://127.0.0.1:5000/get_data");
    let data = await res.json();

    document.getElementById("stairs").innerText =
        "Stairs: " + (data.stairs ? "Available" : "Not Available");

    document.getElementById("traffic").innerText =
        "Traffic: " + data.traffic;

    document.getElementById("wheelchair").innerText =
        "Wheelchair Access: " + (data.wheelchair_access ? "Yes" : "No");
}

async function updateData() {
    let data = {
        stairs: document.getElementById("stairsInput").value === "true",
        traffic: document.getElementById("trafficInput").value,
        wheelchair_access: document.getElementById("wheelInput").value === "true"
    };

    await fetch("http://127.0.0.1:5000/update", {
        method: "POST",
        headers: {
            "Content-Type": "application/json"
        },
        body: JSON.stringify(data)
    });

    fetchData();
}

setInterval(fetchData, 3000);
</script>

</body>
</html>
```
