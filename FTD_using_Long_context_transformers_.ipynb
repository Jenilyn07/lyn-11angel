```python
import mne
import mne.datasets.eegbci

# --- Download and load a sample from PhysioNet EEG Motor Movement/Imagery Database ---

# Define subject and run parameters
subject = 1  # We'll use subject 1 as a sample
runs = [6]   # Run 6: Motor imagery (imagine opening and closing left or right fist)

print(f"Downloading and loading data for Subject {subject}, Run {runs} from PhysioNet BCI dataset...")

try:
    # MNE's function to load EDF files from PhysioNet
    # It automatically handles downloading if not present
    physionet_raw_files = mne.datasets.eegbci.load_data(subject, runs)

    # Read the first raw file (there should only be one for a single run)
    raw_physionet = mne.io.read_raw_edf(physionet_raw_files[0], preload=True)

    print(f"Successfully loaded PhysioNet EEG data from {physionet_raw_files[0]}")

    # The line causing the error 'make_eegbci_browse_raw' has been removed.
    # Inspect properties of the loaded raw object
    print(
"\n--- MNE Raw Object Properties (PhysioNet Dataset) ---"
)
    print(f"Sampling rate: {raw_physionet.info['sfreq']} Hz")
    print(f"Number of channels: {len(raw_physionet.ch_names)}")
    print(f"Channel names (first 5): {raw_physionet.ch_names[:5]} ...")
    print(f"Duration: {raw_physionet.times[-1]:.2f} seconds")

except Exception as e:
    print(f"Error loading PhysioNet EEG data: {e}")
```

```python
import mne

# 1. Prepare Channels
# Clean up channel names by removing any trailing dots
old_ch_names = raw_physionet.info['ch_names']
new_ch_names = [ch.replace('.', '') for ch in old_ch_names]
mapping = dict(zip(old_ch_names, new_ch_names))
raw_physionet.rename_channels(mapping)

# Set channel types. By default, MNE's PhysioNet loader might classify some as MISC.
# The PhysioNet BCI dataset mostly contains EEG channels and one STI 014 for events.
# Let's ensure all relevant channels are set to 'eeg'.
# We will infer EEG channels by checking for 'eeg' in their names, or assuming all are EEG except the stim channel.
ch_types = ['eeg'] * len(raw_physionet.ch_names)

# Identify 'STI 014' if present and set its type as 'stim'
if 'STI 014' in raw_physionet.ch_names:
    stim_idx = raw_physionet.ch_names.index('STI 014')
    ch_types[stim_idx] = 'stim'

# Apply the channel types
raw_physionet.set_channel_types(dict(zip(raw_physionet.ch_names, ch_types)))

# Drop any non-EEG channels that are not needed (e.g. if we had EOG, but for now we focus on EEG and stim)
# For PhysioNet BCI, typically all listed channels are EEG except the stimulus channel.
# No channels need to be dropped at this stage if we keep 'stim' for event extraction.

print("Channel names cleaned and types set.")
print(f"New channel names (first 5): {raw_physionet.ch_names[:5]} ...")
print(f"Channel types: {raw_physionet.get_channel_types(picks='all', unique=True)}")
```

```python
import mne

# 2. Filter Data

# Apply a bandpass filter (0.5 Hz - 30 Hz)
raw_physionet.filter(l_freq=0.5, h_freq=30.0, fir_design='firwin', verbose=False)
print(f"Bandpass filter applied: {raw_physionet.info['lowpass']} Hz - {raw_physionet.info['highpass']} Hz")

# Apply a notch filter at 60 Hz
raw_physionet.notch_filter(freqs=60, fir_design='firwin', verbose=False)
print("Notch filter at 60 Hz applied.")
```

```python
import mne

# 3. Re-reference Data
# Re-reference the EEG data to a common average reference
# The projection=True argument creates a projector that will be applied later
raw_physionet.set_eeg_reference(ref_channels='average', projection=True, verbose=False)
print("EEG data re-referenced to common average.")
```

```python
import mne

# 4. Extract Events and Create Epochs

# Extract events from the raw_physionet annotations
# MNE's PhysioNet BCI dataset uses 'T1' for imagine left fist, 'T2' for imagine right fist
events, event_id = mne.events_from_annotations(raw_physionet, event_id=dict(T1=1, T2=2, T0=3), verbose=False)

print(f"Extracted events: {len(events)} total")
print(f"Event IDs and their mappings: {event_id}")

# Define the specific event IDs you want to epoch (T1 for left fist, T2 for right fist)
# Assuming we are interested in motor imagery events for classification
target_event_id = {'left_fist': event_id['T1'], 'right_fist': event_id['T2']}

# Create mne.Epochs objects
tmin, tmax = -0.2, 0.6  # Time window from -0.2s before event to 0.6s after
baseline = (-0.2, 0) # Baseline correction from -0.2s to 0s

epochs = mne.Epochs(raw_physionet, events, event_id=target_event_id, tmin=tmin, tmax=tmax,
                    baseline=baseline, preload=True, verbose=False)

print(f"Created epochs for events: {target_event_id}")
```

```python
import mne

# 5. Basic Artifact Rejection
# Print the number of generated epochs for each event type
print("\n--- Epochs Information ---")
print(f"Number of epochs for 'left_fist' (T1): {len(epochs['left_fist'])}")
print(f"Number of epochs for 'right_fist' (T2): {len(epochs['right_fist'])}")
print(f"Total number of epochs: {len(epochs)}")

# 6. Inspect Preprocessed Epochs
# Print a summary of the epoched data
print("\n--- Epochs Info ---")
print(epochs.info)

# Check for dropped epochs (if any)
print("\n--- Dropped Epochs Log ---")
print(epochs.drop_log)

# Plot evoked potentials for each event type
# This helps to visually assess the quality of the preprocessed data
print("\nPlotting evoked potentials...")
evt_left_fist = epochs['left_fist'].average()
evt_right_fist = epochs['right_fist'].average()

# Plotting with a specific order for better visualization
mne.viz.plot_compare_evokeds({'left_fist': evt_left_fist, 'right_fist': evt_right_fist}, title='Evoked Potentials for Motor Imagery', picks='eeg')

print("Preprocessing steps completed and inspected.")
```

```python
import numpy as np
import mne

# 1. Define parameters for computing the spectrograms
# Frequencies of interest (e.g., from 4 Hz to 30 Hz for relevant EEG bands)
f_min, f_max = 4, 30
frequencies = np.arange(f_min, f_max + 1, 1) # 1 Hz step

# Number of cycles for the Morlet wavelet transform
# A good starting point is to have a constant number of cycles or increase with frequency
n_cycles = frequencies / 2.  # Example: 0.5 cycles per Hz

# Baseline correction parameters (already defined for epochs, but good to keep consistent)
# tmin_baseline, tmax_baseline = -0.2, 0.0

print(f"Spectrogram parameters defined:\n  Frequencies: {f_min} Hz to {f_max} Hz\n  Number of cycles: {n_cycles[0]:.1f} to {n_cycles[-1]:.1f}")
```

```python
from mne.time_frequency import tfr_morlet

# 2. Iterate through epochs and compute TFR
# Initialize a list to store spectrograms
spectrograms_list = []
epoch_labels = []

print("Computing spectrograms for each epoch...")

# Get event IDs and their integer mappings
# The event_id dict from mne.events_from_annotations can have string keys that are numpy strings
# Convert them to regular strings if necessary for direct comparison
# target_event_id was defined as {'left_fist': 1, 'right_fist': 2}

# Iterate through epochs by index to ensure we get Epochs objects, not just data arrays if iteration is misbehaving
for i in range(len(epochs)):
    current_epoch = epochs[i] # This will be an MNE Epochs object with a single epoch

    # Use the recommended .compute_tfr method
    tfr_data = current_epoch.compute_tfr(method="morlet", freqs=frequencies, n_cycles=n_cycles, use_fft=True,
                                         return_itc=False, decim=3, n_jobs=1, average=False, verbose=False)

    # Extract power data (MNE TFR objects store power in .data attribute)
    # tfr_data.data shape: (n_epochs, n_channels, n_frequencies, n_times)
    # Since we are computing for a single epoch, n_epochs will be 1
    power = tfr_data.data.squeeze(axis=0) # Remove the epoch dimension (now n_channels, n_freq, n_times)

    spectrograms_list.append(power)

    # Get the event label for the current epoch
    # current_epoch.events is an array of (sample, -, event_id)
    current_event_id = current_epoch.events[0, 2]
    label = None
    for k, v in target_event_id.items():
        if v == current_event_id:
            label = k
            break
    epoch_labels.append(label)

print(f"Finished computing {len(spectrograms_list)} spectrograms.")

# Convert the list of spectrograms to a single numpy array
spectrograms_array = np.array(spectrograms_list)
labels_array = np.array(epoch_labels)

print(f"Shape of spectrograms_array: {spectrograms_array.shape} (epochs, channels, frequencies, times)")
print(f"Shape of labels_array: {labels_array.shape}")
```

```python
import numpy as np

# 5. Apply appropriate normalization to the extracted spectrogram power data
# We will use Min-Max scaling to normalize the data to a range of 0 to 1
# This is done across the entire dataset of spectrograms.

# Reshape the spectrograms_array to 2D (epochs * channels * frequencies, times)
# or 1D (all data points) to calculate min/max if scaling globally
# For image-like input, normalizing per-image (or per-channel-per-image) or globally is common.
# Let's normalize globally across all spectrograms to keep relative differences.

min_val = np.min(spectrograms_array)
max_val = np.max(spectrograms_array)

# Avoid division by zero if all values are the same (unlikely but good practice)
if (max_val - min_val) == 0:
    normalized_spectrograms = np.zeros_like(spectrograms_array)
else:
    normalized_spectrograms = (spectrograms_array - min_val) / (max_val - min_val)

print(f"Spectrograms normalized to range [{np.min(normalized_spectrograms):.2f}, {np.max(normalized_spectrograms):.2f}]")
print(f"Shape of normalized_spectrograms: {normalized_spectrograms.shape}")
print(f"Shape of labels_array: {labels_array.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras.layers import (
    Input, Conv2D, Reshape, Dense, LayerNormalization,
    MultiHeadAttention, Dropout, GlobalAveragePooling1D, MaxPooling2D, Add
)
from tensorflow.keras.models import Model
import numpy as np # Added numpy import

# Transpose spectrograms to channels_last format (epochs, frequencies, times, channels)
# Explicitly cast to float32 for TensorFlow compatibility. This ensures INPUT_SHAPE can be defined.
X_train_transposed = np.transpose(X_train, (0, 2, 3, 1)).astype(np.float32)

# The input shape now needs to be (frequencies, times, channels) as data was transposed
INPUT_SHAPE = X_train_transposed.shape[1:] # (frequencies, times, channels)
NUM_CLASSES = num_classes # Already determined from label encoding

def build_cnn_transformer_model(input_shape, num_classes):
    inputs = Input(shape=input_shape)

    # CNN Feature Extractor
    # Removed data_format='channels_first' as input is now channels_last (NHWC)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)

    # Reshape for Transformer: (batch_size, sequence_length, embedding_dim)
    # The output of CNN is now (batch_size, height, width, filters) = (batch_size, H_out, W_out, C_out)
    cnn_output_shape = x.shape # (None, H_out, W_out, C_out)

    # sequence_length = H_out * W_out
    # embedding_dim = C_out (filters)
    sequence_length = cnn_output_shape[1] * cnn_output_shape[2]
    embedding_dim = cnn_output_shape[3]

    # Reshape directly to (batch_size, sequence_length, embedding_dim)
    x = Reshape((sequence_length, embedding_dim))(x)

    # Transformer Encoder Block
    # Define hyperparameters for transformer
    num_heads = 4  # Number of attention heads
    ff_dim = 128   # Hidden layer size in feed forward network inside transformer

    # Multi-head attention layer
    attn_output = LayerNormalization(epsilon=1e-6)(x)
    attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=x.shape[-1])(attn_output, attn_output)
    attn_output = Dropout(0.3)(attn_output)
    x = Add()([x, attn_output]) # Residual connection

    # Feed Forward Network
    ffn_output = LayerNormalization(epsilon=1e-6)(x)
    ffn_output = Dense(ff_dim, activation='relu')(ffn_output)
    ffn_output = Dense(x.shape[-1])(ffn_output)
    ffn_output = Dropout(0.3)(ffn_output)
    x = Add()([x, ffn_output]) # Residual connection

    # Classification Head
    x = GlobalAveragePooling1D()(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model

# Build the model with the corrected input shape
model = build_cnn_transformer_model(INPUT_SHAPE, NUM_CLASSES)

# Compile the model
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
loss_fn = tf.keras.losses.CategoricalCrossentropy()
metrics = ['accuracy']

model.compile(optimizer=optimizer, loss=loss_fn, metrics=metrics)

print("CNN-Transformer Hybrid Model rebuilt and recompiled successfully with corrected input format.")
model.summary()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import (
    Input, Conv2D, Reshape, Dense, LayerNormalization,
    MultiHeadAttention, Dropout, GlobalAveragePooling1D, MaxPooling2D, Add
)
from tensorflow.keras.models import Model

# The input shape now needs to be (frequencies, times, channels) as data was transposed
INPUT_SHAPE = X_train_transposed.shape[1:] # (frequencies, times, channels)
NUM_CLASSES = num_classes # Already determined from label encoding

def build_cnn_transformer_model(input_shape, num_classes):
    inputs = Input(shape=input_shape)

    # CNN Feature Extractor
    # Removed data_format='channels_first' as input is now channels_last (NHWC)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    x = MaxPooling2D((2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D((2, 2))(x)

    # Reshape for Transformer: (batch_size, sequence_length, embedding_dim)
    # The output of CNN is now (batch_size, height, width, filters) = (batch_size, H_out, W_out, C_out)
    cnn_output_shape = x.shape # (None, H_out, W_out, C_out)

    # sequence_length = H_out * W_out
    # embedding_dim = C_out (filters)
    sequence_length = cnn_output_shape[1] * cnn_output_shape[2]
    embedding_dim = cnn_output_shape[3]

    # Reshape directly to (batch_size, sequence_length, embedding_dim)
    x = Reshape((sequence_length, embedding_dim))(x)

    # Transformer Encoder Block
    # Define hyperparameters for transformer
    num_heads = 4  # Number of attention heads
    ff_dim = 128   # Hidden layer size in feed forward network inside transformer

    # Multi-head attention layer
    attn_output = LayerNormalization(epsilon=1e-6)(x)
    attn_output = MultiHeadAttention(num_heads=num_heads, key_dim=x.shape[-1])(attn_output, attn_output)
    attn_output = Dropout(0.3)(attn_output)
    x = Add()([x, attn_output]) # Residual connection

    # Feed Forward Network
    ffn_output = LayerNormalization(epsilon=1e-6)(x)
    ffn_output = Dense(ff_dim, activation='relu')(ffn_output)
    ffn_output = Dense(x.shape[-1])(ffn_output)
    ffn_output = Dropout(0.3)(ffn_output)
    x = Add()([x, ffn_output]) # Residual connection

    # Classification Head
    x = GlobalAveragePooling1D()(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model

# Build the model with the corrected input shape
model = build_cnn_transformer_model(INPUT_SHAPE, NUM_CLASSES)

# Compile the model
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)
loss_fn = tf.keras.losses.CategoricalCrossentropy()
metrics = ['accuracy']

model.compile(optimizer=optimizer, loss=loss_fn, metrics=metrics)

print("CNN-Transformer Hybrid Model rebuilt and recompiled successfully with corrected input format.")
model.summary()
```

```python
import gradio as gr
import numpy as np

description_text = """
EEG Motor Imagery Classification Demo.

• A sample EEG spectrogram image is shown separately for visualization.
• Prediction is performed using the normalized spectrogram tensor (.npy file).
• The model classifies between 'left_fist' and 'right_fist'.
"""

def predict_from_npy(file):
    """
    file: uploaded normalized_spectrograms.npy
    """

    # Load the .npy spectrogram data
    data = np.load(file.name)

    # Select ONE sample (the first one) from the loaded data
    # Original data shape: (epochs, channels, frequencies, times) -> e.g., (15, 64, 27, 43)
    single_epoch_data = data[0] # Shape: (64, 27, 43) -> (channels, frequencies, times)

    # Transpose to model's expected input format (frequencies, times, channels)
    # The model was trained with input shape (frequencies, times, channels)
    spectrogram_for_model = np.transpose(single_epoch_data, (1, 2, 0)) # Shape: (27, 43, 64)

    # Add batch dimension (model expects (batch_size, frequencies, times, channels))
    spectrogram_for_model = np.expand_dims(spectrogram_for_model, axis=0) # Shape: (1, 27, 43, 64)

    # Model prediction
    predictions = model.predict(spectrogram_for_model, verbose=0) # verbose=0 to suppress console output during inference

    return {
        "left_fist": float(predictions[0][0]),
        "right_fist": float(predictions[0][1])
    }

iface = gr.Interface(
    fn=predict_from_npy,
    inputs=gr.File(label="Upload normalized EEG spectrogram (.npy)"),
    outputs=gr.Label(num_top_classes=2, label="Prediction"),
    title="EEG Motor Imagery Classification",
    description=description_text,
    allow_flagging="never"
)

print("Launching Gradio application...")
iface.launch(share=False)
```
